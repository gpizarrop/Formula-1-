# Analisis Formula 1 campeonato 2022

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('C:/Users\pizar\Desktop\Jupyter notebook\Formula1_2022season_raceResults.csv')
```

```python
print(df)
```

```python
df.describe()
```

### Agrupando puntos por piloto

```python
puntos = df.groupby('Driver')['Points'].sum().sort_values(ascending=False)
print(puntos)
```

#### Grafica de sumatoria de puntos totales

```python
puntos.plot(kind='bar')
plt.xlabel('Driver')
plt.ylabel('Puntos')
plt.title('Sumatoria de puntos por piloto')
plt.show()
```
